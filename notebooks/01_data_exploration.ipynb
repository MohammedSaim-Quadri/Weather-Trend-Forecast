```python
# replace extreme values with Nan in air quality features
air_quality_features = [
    "air_quality_Carbon_Monoxide","air_quality_Ozone", 
    "air_quality_Nitrogen_dioxide", "air_quality_Sulphur_dioxide",
    "air_quality_PM2.5", "air_quality_PM10"
]
```

```python
df[air_quality_features] = df[air_quality_features].replace(-9999, np.nan)
```

```python
# cap extreme values using 99th percentile
cap_features = ["wind_mph", "wind_kph", "gust_mph", "gust_kph", "pressure_mb"]
for feature in cap_features:
    cap_value = df[feature].quantile(0.99)
    df[feature] = np.where(df[feature] > cap_value, cap_value, df[feature])
```

```python
# convert epoch time to datetime format
df["last_updated"] = pd.to_datetime(df["last_updated_epoch"], unit="s")
df.drop(columns=["last_updated_epoch"], inplace=True) 
```

```python
# check zero visibility cases(potential sensor issues)
zero_visibility_count = (df["visibility_km"] == 0).sum()
print(f"Zero visibility instances: {zero_visibility_count}")
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=air_quality_features, inplace=True)
```

```python
df.isnull().sum()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical Columns:", categorical_cols)
```

```python
# frequency encoding for high cardinality features
"""
country: 210 unique values
location_name: 248 unique values
timezone: 190 unique values
last_updated: 10457 unique values
condition_text: 47 unique values
wind_direction: 16 unique values
sunrise: 456 unique values
sunset: 534 unique values
moonrise: 1441 unique values
moonset: 1441 unique values
moon_phase: 8 unique values
"""
high_cardinality_features = ['country', 'location_name', 'timezone']

for col in high_cardinality_features:
    freq_map = df[col].value_counts(normalize=True)
    df[col] = df[col].map(freq_map)
```

```python
# one hot encoding for condition_text
df = pd.get_dummies(df, columns=['condition_text'], drop_first=True)
```

```python
df['wind_direction'].head()
```

```python
# convert wind_direction to degrees
wind_map = {
    'N': 0, 'NNE': 22.5, 'NE': 45, 'ENE': 67.5, 'E': 90, 'ESE': 112.5, 'SE': 135, 'SSE': 157.5,
    'S': 180, 'SSW': 202.5, 'SW': 225, 'WSW': 247.5, 'W': 270, 'WNW': 292.5, 'NW': 315, 'NNW': 337.5
}
df['wind_direction'] = df['wind_direction'].map(wind_map)
```

```python
from sklearn.preprocessing import LabelEncoder
df['moon_phase'] = LabelEncoder().fit_transform(df['moon_phase'])
```

```python
def time_to_minutes(time_str):
    # Convert time to datetime format to handle AM/PM
    time_obj = pd.to_datetime(time_str, format='%I:%M %p', errors='coerce')
    
    if pd.isna(time_obj):  # Handle errors
        return None
    
    return time_obj.hour * 60 + time_obj.minute

# Convert each time column
time_cols = ['sunrise', 'sunset', 'moonrise', 'moonset']
for col in time_cols:
    df[col] = df[col].apply(time_to_minutes)

# Check if any nulls were created
print(df[time_cols].isnull().sum())
```

```python
print(df[df['moonrise'].isna()][['moonrise']].head(10))
print(df[df['moonset'].isna()][['moonset']].head(10))
```

```python
missing_percent = df[['moonrise', 'moonset']].isna().sum() / len(df) * 100
print(missing_percent)
```

```python
df.dropna(subset=['moonrise', 'moonset'], inplace=True)
```

```python
# extract datetime features from last_updated
df['last_updated'] = pd.to_datetime(df['last_updated'])
df['hour'] = df['last_updated'].dt.hour
df['day'] = df['last_updated'].dt.day
df['month'] = df['last_updated'].dt.month
df['year'] = df['last_updated'].dt.year
```
